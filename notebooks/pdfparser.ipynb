# PDF Paragraph Parser

Extracts paragraph-level text from bank annual report PDFs using PyMuPDF and saves a clean Excel file ready for ClimateBERT classification.

**Input:** One PDF following the naming convention `BankName_Year_ReportType.pdf`
**Output:** `results/parsed/bankname_year_reporttype_YYYY-MM-DD.xlsx` — one row per paragraph

## 4. Upload PDF

```python
from google.colab import files
import shutil, os

uploaded = files.upload()
filename = list(uploaded.keys())[0]

os.makedirs("/content/thesis/data/pdfs", exist_ok=True)
shutil.move(filename, f"/content/thesis/data/pdfs/{filename}")
print(f"Ready: {filename}")
```

## 5. Run PDF Parser

```python
from pdf_parser import parse_pdf

df, output_file = parse_pdf(f"/content/thesis/data/pdfs/{filename}")
```

## 6. Inspect Results

```python
print(f"Total paragraphs: {len(df)}")
print(f"Word count range: {df['word_count'].min()} – {df['word_count'].max()}")
print(f"Mean word count: {df['word_count'].mean():.0f}")
print(df.head(10))
```
